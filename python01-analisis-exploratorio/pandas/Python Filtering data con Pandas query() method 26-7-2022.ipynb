Sintaxis: DataFrame.query(expr, inplace=False, **kwargs)

Parámetros:

expr: Expresión en forma de cadena para filtrar datos.

inplace: Realice cambios en el marco de datos original si es Verdadero

kwargs: Otros argumentos de palabras clave.

Tipo de devolución: marco de datos filtrados

```python
import pandas as pd

# Cargamos los datos
data = pd.read_csv("employees.csv")

# Reemplazo los especios en blanco por '_'
data.columns =[column.replace(" ", "_") for column in data.columns]

# Filtramos por el método Query
data.query('Senior_Management == True', inplace = True)

#Mostramos el data set
data
```

Sintaxis:
DataFrame.where(cond, other=nan, inplace=False, axis=None, level=None, errors=’raise’, try_cast=False, raise_on_error=None)
 

Parameteros:

cond: una o más condiciones para verificar el marco de datos.

otro: reemplaza las filas que no cumplen la condición con el objeto definido por el usuario, el valor predeterminado es NaN

inplace: valor booleano, realiza cambios en el propio marco de datos si es verdadero

eje: eje a comprobar (fila o columnas)

```python
import pandas as pd

data = pd.read_csv("nba.csv")

# ordenamos
data.sort_values("Team", inplace = True)

# especificamos la condición
filter = data["Team"]=="Atlanta Hawks"

# Filtramos
data.where(filter, inplace = True)

# Mostramos
data

#Cuando no se cumple la condición se muestra Nan
```

```python
import pandas as pd

data = pd.read_csv("nba.csv")

data.sort_values("Team", inplace = True)

filter1 = data["Team"]=="Atlanta Hawks"

filter2 = data["Age"]>24

# filtramos por más de una condición
data.where(filter1 & filter2, inplace = True)

# Mostramos
data
```
